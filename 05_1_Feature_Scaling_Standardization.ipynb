```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv('Bengaluru_House_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(x='bath',y='price', data=df)
```

```python
scaler = StandardScaler()
```

```python
df['Price_scale']= scaler.fit_transform(df[['price']])
```

```python
df.head()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
sns.scatterplot(x='price', y='bath', data=df)

plt.subplot(1,2,2)
sns.scatterplot(x='Price_scale',y='bath', data=df)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(df['price'], ax=ax1)
sns.kdeplot(df['bath'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(df['Price_scale'], ax=ax2)
sns.kdeplot(df['bath'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Price Distribution Before Scaling')
sns.kdeplot(df['price'], ax=ax1)

# after scaling
ax2.set_title('Price Distribution After Standard Scaling')
sns.kdeplot(df['Price_scale'], ax=ax2)
plt.show()
```
