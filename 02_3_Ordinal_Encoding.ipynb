```python
ordinal_cols= ['Job Satisfaction','Job Level', 'Company Size', 'Company Reputation', 'Employee Recognition', 'Education Level']
```

```python
oe = OrdinalEncoder(categories=[[ 'Low','Medium', 'High', 'Very High'],
                                ['Entry','Mid', 'Senior'],
                                ['Small','Medium', 'Large'],
                                ['Poor','Fair','Good', 'Excellent'],
                                ['Low','Medium', 'High', 'Very High'],
                                ['High School', 'Associate Degree','Bachelor’s Degree', 'Master’s Degree', 'PhD']]) 
```

```python
df[ordinal_cols] = oe.fit_transform(df[ordinal_cols])
df.head()
```
