```python
upper_limit = df['price'].quantile(0.95)
upper_limit
```

```python
lower_limit = df['price'].quantile(0.05)
lower_limit
```

```python
df[(df['price']>= 324) | (df['price']<=31.5)]
```

```python
new_df = df[(df['price']<= 324) & (df['price']>=31.5)]
```

# Capping & Winsorization

```python
df['price']= np.where(
    df['price']>= upper_limit,
    upper_limit,
    np.where(
        df['price']<= lower_limit,
        lower_limit,
        df['price']
    )
)
```
