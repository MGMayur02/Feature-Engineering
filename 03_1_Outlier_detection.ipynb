# Capping

```python
upper_limit = df['price'].mean() +3*df['price'].std()
lower_limit = df['price'].mean() - 3 * df['price'].std()
```

```python
upper_limit
```

```python
lower_limit
```

```python
df['price'] = np.where(
    df['price']> upper_limit,
    upper_limit,
    np.where(
        df['price']<lower_limit,
        lower_limit,
        df['price']
    )
)
```
