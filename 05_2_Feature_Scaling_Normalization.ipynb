# MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
x = scaler.fit_transform(df)
```

```python
x= pd.DataFrame(x, columns=df.columns)
```

```python
x.head()
```

```python
fig, (ax1, ax2) =plt.subplots(ncols= 2, figsize=(12,5))

ax1.scatter(df['alcohol'], df['citric acid'])
ax1.set_title('Before')
ax2.scatter(x['alcohol'], x['citric acid'])
ax2.set_title('After')
```

```python
fig, (ax1, ax2) =plt.subplots(ncols= 2, figsize=(12,5))
ax1.set_title('Before')
sns.kdeplot(df['alcohol'], ax=ax1)
sns.kdeplot(df['citric acid'], ax=ax1)
ax2.set_title('After')
sns.kdeplot(x['alcohol'],ax=ax2)
sns.kdeplot(x['citric acid'],ax=ax2)
```
