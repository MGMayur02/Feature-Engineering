```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import scipy.stats as stats
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
```

```python
df.head()
```

```python
df['price'].isna().sum()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,2,1)
sns.histplot(x='price',data= df, kde=True)
plt.title('Price')

plt.subplot(1,2,2)
stats.probplot(df['price'], dist='norm', plot=plt)
plt.title('Price')
```

# Log Tranformer

```python
from sklearn.preprocessing import FunctionTransformer 
```

```python
trans = FunctionTransformer(func=np.log1p)
```

```python
df['price_trans'] = trans.fit_transform(df[['price']])
```

```python
df.head()
```

```python
plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
sns.histplot(x='price',data= df, kde=True)
plt.title('Price Before log',  fontsize=15)

plt.subplot(2,2,2)
stats.probplot(df['price'], dist='norm', plot=plt)
plt.title('Price Before log', fontsize=15)

plt.subplot(2,2,3)
sns.histplot(x='price_trans',data= df, kde=True)
plt.title('Price After log', fontsize=15)

plt.subplot(2,2,4)
stats.probplot(df['price_trans'], dist='norm', plot=plt)
plt.title('Price After log', fontsize=15)
```

# Reciprocal Transformal

```python
trans2 = FunctionTransformer(func=np.reciprocal)
```

```python
df['price_trans_reci'] = trans2.fit_transform(df[['price']])
```

```python
df.head()
```

```python
plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
sns.histplot(x='price',data= df, kde=True)
plt.title('Price Before Reciprocal',  fontsize=15)

plt.subplot(2,2,2)
stats.probplot(df['price'], dist='norm', plot=plt)
plt.title('Price Before Reciprocal', fontsize=15)

plt.subplot(2,2,3)
sns.histplot(x='price_trans_reci',data= df, kde=True)
plt.title('Price After Reciprocal', fontsize=15)

plt.subplot(2,2,4)
stats.probplot(df['price_trans_reci'], dist='norm', plot=plt)
plt.title('Price After Reciprocal', fontsize=15)
```

# Square Transformer

```python
trans3 = FunctionTransformer(func=np.square)
```

```python
df['price_trans_square']=trans3.fit_transform(df[['price']])
```

```python
df.head()
```
